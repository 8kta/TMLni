# Proyect 1 ML

Se pide hacer un modelo que prediga los precios de la finca raiz dada una base de datos que fue tomada de https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data se pide una proyeccion utilizando los metodos de regresion tales como el lineal, Lasso y Ridge Model.


```python
from scipy import stats
import numpy as np
import pandas as pd 
import matplotlib
import matplotlib.pyplot as plt
import math
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder    #Se utiliza para cambiar valores no numericos y codificarlos numericamente 
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.linear_model import Lasso 
from sklearn.linear_model import Ridge
from sklearn.linear_model import LinearRegression
from collections import defaultdict
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

Veamos que datos tenemos.

```python
train
```

## Preparacion de datos
Dado que tenemos datos de muchas clases queremos hacer una limpieza basica de datos, cambiamos los valores NaN y los valores no numerico por numericos, para esto primero unimos los datos para codificarlos al tiempo y que tengan los mismos valores numericos acorde con su dato clasico.
Ademas separamos los datos de trin para dejar saleprice separado para nuestro modelo de regresion.

```python
test = test.fillna(0)
train = train.fillna(0)

saleprice = train[['SalePrice']]
train1 = train.drop(['SalePrice'],axis=1)
union = pd.concat([train1, test])
```

Tambien quitaremos la columna 1 que no esta relacionada con los precios

```python
union1 = union.drop(['Id'],axis=1)
```

Ahora si quitamos transformamos los valores no numericos.

```python
pre = preprocessing.LabelEncoder()
mask = union1.dtypes.eq(np.object)
union1.loc[:, mask] = union1.loc[:, mask].astype(str).apply(pre.fit_transform)
```

Por ultimo separamos de nuevo train y test para realizar las regresiones.

```python
train1 = union1.iloc[:1460,:] 
test1= union1.iloc[1460:,:]
```

# Modelos de regresion
## Regresion lineal

Subiremos los datos relaes de los precios para comparar la prediccion con los reales.

```python
Real = pd.read_csv("sample_submission.csv")
```

```python
linear_model = LinearRegression(normalize=False)
linear_model.fit(train1,saleprice)
```

```python
plt.plot(test1.index, linear_model.predict(test1),'b',alpha=0.6, label='Prediction')
plt.plot(test1.index, Real,'r--', label='Data')
plt.legend(loc='best', frameon=False)
plt.xlabel('Houses')
plt.ylabel('price')
plt.show()
```

## Modelo Lasso

```python
lasso_model = Lasso(alpha=0.5,normalize=True, max_iter=1e6)
lasso_model.fit(train1,saleprice)
```

```python
plt.plot(test1.index, lasso_model.predict(test1),'b',alpha=0.6, label='Prediction')
plt.plot(test1.index, Real,'r--', label='Data')
plt.legend(loc='best', frameon=False)
plt.xlabel('LotArea')
plt.ylabel('y')
plt.show()
```

## Modelo Ridge 

```python
ridge_model = Ridge(alpha=0.5)
ridge_model.fit(train1, saleprice) 
```

```python
plt.plot(test1.index, ridge_model.predict(test1),'b',alpha=0.6, label='Prediction')
plt.plot(test1.index, Real,'r--', label='Data')
plt.legend(loc='best', frameon=False)
plt.xlabel('LotArea')
plt.ylabel('y')
plt.show()
```

Con lo que vemos que los tres modelos son bastante similares, sin embargo utilizaremos el modelo de Lasso, con lo que la prediccion para el test dado queda.

```python
f=lasso_model.predict(test1)
i= test[['Id']]
i['p']=f
i
```
